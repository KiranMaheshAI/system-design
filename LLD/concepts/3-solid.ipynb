## SOLID Principles

The SOLID principles provide a blueprint for writing code that’s easy to adjust, extend, and maintain over time.

### S: Single responsibility Principle:

A class should have one, and only one, reason to change. This means that a class must have only one responsibility.

When a class performs just one task, it contains a small number of methods and member variables making them more usable and easier to maintain.

If a class has multiple responsibilities, it becomes harder to understand, maintain, and modify and increases the potential for bugs because changes to one responsibility could affect the others.

```python
# Code Example: Which violates the Single Responsibility Principle?
# Imagine you have a class called UserManager that handles user authentication, user profile management, and email notifications.

class UserManager:
    def authenticate_user(self, username, password):
        # Authenticate user
        pass

    def update_user_profile(self, user_id, new_name, new_email):
        # Update user profile
        pass

    def send_email_notification(self, user_id, message):
        # Send email notification
        pass
```

```python
# To adhere to the SRP, we can split this class into three separate classes, each with a single responsibility:

class UserAuthenticator:
    def authenticate_user(self, username, password):
        # Authenticate user
        pass

class UserProfileManager:
    def update_user_profile(self, user_id, new_name, new_email):
        # Update user profile
        pass

class EmailNotificationManager:
    def send_email_notification(self, user_id, message):
        # Send email notification

# Now, each class has a single, well-defined responsibility. Changes to user authentication won't affect the email notification logic, 
# and vice versa, improving maintainability and reducing the risk of unintended side effects.
```

### O: Open/Closed Principle (OCP)

Software entities (classes, modules, functions, etc.) should be open for extension, but closed for modification.

This means the design of a software entity should be such that you can introduce new functionality or behavior without modifying the existing code since changing the existing code might introduce bugs.

```python
# Code Example: which violates the Open/Closed Principle?
# Imagine you have a class called Shape that has a method to calculate the area of a shape.

class Shape:
    def calculate_area(self, shape):
        if shape == "circle":
            return 3.14 * self.radius ** 2
        elif shape == "rectangle":
            return self.width * self.height
        else:
            return 0
    def calculate_perimeter(self, shape):
        if shape == "circle":
            return 2 * 3.14 * self.radius
        elif shape == "rectangle":
            return 2 * (self.width + self.height)
        else:
            return 0

# If we want to add support for a new shape, like a triangle, we would have to modify the calculate_area and calculate_perimeter methods, 
# violating the Open/Closed Principle.
```

```python
# To adhere to the OCP, we can use inheritance to extend the Shape class to include the new shape.

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    
    @abstractmethod
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def calculate_area(self):
        return 3.14 * self.radius ** 2
    def calculate_perimeter(self):
        return 2 * 3.14 * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def calculate_area(self):
        return self.width * self.height
    def calculate_perimeter(self):
        return 2 * (self.width + self.height)

# We can add new shapes likes triangle without modifying existing code.
class Triangle(Shape):
    pass
```

### L: Liskov Substitution Principle (LSP)

Objects of a superclass should be replaceable with objects of its subclasses without affecting the correctness of the program.

This means if you have a base class and a derived class, you should be able to use instances of the derived class wherever instances of the base class are expected, without breaking the application.

```python
# Code Example: which violates the Liskov Substitution Principle?
# Imagine you have a class called Vehicle that has a method to calculate the fuel consumption of a vehicle.

class Vehicle:
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Starting car engine")

class Motorcycle(Vehicle):
    def start_engine(self):
        pass

# In this example, the Bicycle class violates the LSP because it provides an implementation for the start_engine method, 
# which doesn't make sense for a bicycle.
```

```python
# to adhere to the LSP

class Vehicle:
    def start_engine(self):
        raise NotImplementedError("Subclasses must implement this method")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("Starting motorcycle engine")

# Here, we've replaced the start_engine method with a more general start method in the base class Vehicle. 
# The Car class implements the start method to start the engine, while the Bicycle class implements the start method 
# to indicate that the rider is pedaling.

# Now, instances of Car and Bicycle can be safely substituted for instances of Vehicle without any unexpected behavior or errors.
```

### I: Interface Segregation Principle (ISP)

No client should be forced to depend on interfaces they don't use.

By segregating interfaces into smaller, more specific ones, clients only depend on the methods they actually need, promoting loose coupling and better code organization.

```python
# Code Example: which violates the Interface Segregation Principle?

class MediaPlayer:
    def play_audio(self):
        raise NotImplementedError("Subclasses must implement this method")

    def play_video(self):
        raise NotImplementedError("Subclasses must implement this method")
    
    def stop_audio(self):
        raise NotImplementedError("Subclasses must implement this method")

    def stop_video(self):
        raise NotImplementedError("Subclasses must implement this method")
    
    def adjust_volume(self, volume):
        raise NotImplementedError("Subclasses must implement this method")

    def adjust_brightness(self, brightness):
        raise NotImplementedError("Subclasses must implement this method")

# In this case, any class that implements the MediaPlayer interface would be forced to implement all the methods, even if it doesn't need them.
```

```python
# To adhere to the ISP
class AudioPlayer:
    def play_audio(self):
        raise NotImplementedError("Subclasses must implement this method")

    def stop_audio(self):
        raise NotImplementedError("Subclasses must implement this method")

    def adjust_volume(self, volume):
        raise NotImplementedError("Subclasses must implement this method")

class VideoPlayer:
    def play_video(self):
        raise NotImplementedError("Subclasses must implement this method")
    def stop_video(self):
        raise NotImplementedError("Subclasses must implement this method")
    def adjust_brightness(self, brightness):
        raise NotImplementedError("Subclasses must implement this method")

# Now, clients can depend on specific interfaces without being forced to implement unnecessary methods.
class MP3Player(AudioPlayer):
    def play_audio(self, audio_file):
        print(f"Playing MP3 file: {audio_file}")

class MP4Player(VideoPlayer):
    def play_video(self, video_file):
        print(f"Playing MP4 file: {video_file}")

# By segregating the interfaces, each class only needs to implement the methods it actually requires. 
# This not only makes the code more maintainable but also prevents clients from being forced to depend on methods they don't use.

class MultiMediaPlayer(AudioPlayer, VideoPlayer):   
    # Implement both interfaces
    pass

# Now, the MultiMediaPlayer class can implement both the AudioPlayer and VideoPlayer interfaces, 
# allowing it to play both audio and video files.
```

### D: Dependency Inversion Principle (DIP)

High-level modules should not depend on low-level modules; both should depend on abstractions.

This means that a particular class should not depend directly on another class, but on an abstraction (interface) of this class.

```python
# Code Example: which violates the Dependency Inversion Principle?

class GmailClient:
    def send_email(self, to, subject, body):
        # Write code to send email using Gmail

class OutlookClient:
    def send_email(self, to, subject, body):
        # Write code to send email using Outlook

# In this case, the GmailClient and OutlookClient classes are tightly coupled to each other, 
# and if we want to switch to a different email service, we would need to modify the EmailSender class.

class EmailService:
    def __init__(self):
        self.gmail_client = GmailClient()
        self.outlook_client = OutlookClient()

    def send_email(self, to, subject, body):
        self.gmail_client.send_email(to, subject, body)
        self.outlook_client.send_email(to, subject, body)
```

```python
# To adhere to the DIP, we can introduce an abstraction (interface) for the email service.
class EmailClient:
    def send_email(self, to, subject, body):
        raise NotImplementedError("Subclasses must implement this method")

class GmailClient(EmailClient):
    def send_email(self, to, subject, body):
        # Write code to send email using Gmail

class OutlookClient(EmailClient):
    def send_email(self, to, subject, body):
        # Write code to send email using Outlook

class EmailService:
    def __init__(self, email_client):
        self.email_client = email_client

    def send_email(self, to, subject, body):
        self.email_client.send_email(to, subject, body)

# Now, the EmailService class depends on the EmailClient interface, not the specific implementations of GmailClient or OutlookClient. 
# This allows us to easily switch between different email services by changing the EmailClient implementation without affecting the EmailService class.
```
