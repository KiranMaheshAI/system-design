## Low Level Design (LLD)

Low Level Design refers to the process of building a complete application structure around one or multiple Data Structures and Algorithms (DSA) concepts. While DSA acts as the "brain" of an application by providing logic for isolated problems, LLD serves as the "skeleton," providing the internal framework and architecture that holds the application together.

### Difference between HLD, LLD and DSA:

The fundamental differences between Data Structures and Algorithms (DSA), Low Level Design (LLD), and High Level Design (HLD) lie in their focus, the problems they solve, and their role in building a scalable application. While all three are necessary to create a software application, they operate at different layers of the development process,.

#### Data Structures and Algorithms (DSA):
DSA is described as the "brain" of an application. Its primary focus is on solving isolated, specific problems through efficient logic.
• Purpose: It deals with the algorithmic implementation required to make a feature work. For example, in a ride-sharing app, DSA would be used to find the shortest path between a source and destination (using Dijkstra’s algorithm) or to assign the closest rider to a user (using a Priority Queue),.
• Scope: DSA is a tool used within the LLD phase to solve logical challenges. It does not account for how different parts of the code interact or how the system will scale.
#### Low Level Design (LLD)
If DSA is the brain, LLD is the "skeleton" of the application. It focuses on the internal structure of the code and how different components are organized.
• Purpose: LLD identifies the objects and entities within an application (such as "User," "Rider," or "Payment") and defines the relationships and interactions between them.
• Core Focus Areas:
    ◦ Scalability: Writing code that can handle a growing number of users without breaking.
    ◦ Maintainability: Ensuring the code is easily debuggable and that new features can be added without breaking existing ones.
    ◦ Reusability: Creating a "plug and play" model where code is not "tightly coupled" to one specific application but can be reused elsewhere (e.g., a generic payment gateway or notification system),.
• Security: It addresses internal concerns like data security, ensuring sensitive information like phone numbers is handled correctly between entities,.
#### High Level Design (HLD)
HLD focuses on the overall system architecture and the external infrastructure rather than the internal code,.
• Purpose: It deals with the high-level components of a system, such as system design diagrams, and rarely involves writing actual code.
• Key Decisions:
    ◦ Tech Stack: Deciding which languages or frameworks to use (e.g., Java Spring Boot).
    ◦ Databases: Choosing between SQL, NoSQL, or hybrid database models,.
    ◦ Server Scaling: Managing how servers will expand or contract to handle bulk traffic.
    ◦ Cost Optimization: Ensuring the system runs efficiently to minimize the cost of services like AWS

## Low-Level Design (Software) 
What it is: The granular phase of software development that bridges High-Level Design (HLD) and actual coding, detailing how components work, interact, and use specific algorithms and data structures.

Key Focus Areas: Classes, objects, methods, interfaces, data structures, and design patterns (Creational, Structural, Behavioral).

Purpose: To create clean, flexible, maintainable, and testable code.

Process: Involves defining component architecture, data models, APIs, and writing UML diagrams, often practiced with real-world examples like designing a parking lot or booking system.

### Three Pillars of LLD:
Here are three pillars of LLD.
1. Scalability
2. Maintainability
3. Reusability
